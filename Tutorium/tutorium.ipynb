```python
import pandas as pd
import yfinance as yf
```

```python
aapl = yf.download(
    tickers="AAPL",
    start="2020-01-01",
    end="2022-01-01",
    interval="1d",
    auto_adjust=False,
    progress=False
)
aapl.dtypes
missing_counts = aapl.isna().sum().sort_values(ascending=False)
missing_pct = (aapl.isna().mean() * 100).sort_values(ascending=False)
pd.DataFrame({"missing_count": missing_counts, "missing_pct": missing_pct})
```

```python
data = aapl.dropna()
```

```python
x = data.loc[:, [("Open", "AAPL"), ("High", "AAPL"), ("Low", "AAPL")]]
y = data.loc[:, [("Close", "AAPL")]]
```

```python
split_index = int(len(data) * 0.8)
split_index

x_train,x_test = x[:split_index]  ,x[split_index:]
print(x_train)
print(x_test)
```

```python
y_train,y_test = y[:split_index]  ,y[split_index:]
print(y_train)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
type(scaler)
```

```python
#use only train data
scaler = scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
data["Close_lag1"]= data[("Close", "AAPL")].shift(1)
data.dropna(inplace=True)
data
```

```python
corr = data[ [("Open", "AAPL"), ("High", "AAPL"), ("Low", "AAPL"), ("Close", "AAPL"),("Volume", "AAPL")]].corr()
corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
hm = sns.heatmap(corr, annot=True,cmap="coolwarm"     )
plt.title("Correlation Heatmap")
plt.show()
```
